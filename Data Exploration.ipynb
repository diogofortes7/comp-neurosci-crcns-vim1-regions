```python

# To get all V1 voxel responses in the training data set:
import pandas as pd
import tables,numpy
f = tables.open_file('EstimatedResponses.mat')
f.list_nodes # Show data file structure with all variables available
Dat = f.get_node('/dataTrnS1')[:] #get the response data for all voxels and all images #TrnS1 - 5 training runs for subject 1
ROI = f.get_node('/roiS1')[:].flatten() #get the ROI info for subject 1
V1idx = numpy.nonzero(ROI==1)[0] #returns the indices for when ROI == 1 (V1 region)
V1resp = Dat[:,V1idx] # get all rows, but only those voxels (columns) corresponding to a region of interest (V1)
```
